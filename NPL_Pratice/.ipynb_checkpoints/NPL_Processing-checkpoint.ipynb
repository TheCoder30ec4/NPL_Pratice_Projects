```python
import nltk
from nltk.corpus import twitter_samples
import matplotlib.pyplot as plt
import random
```

```python
nltk.download('twitter_samples')
```

```python
all_positive_tweets = twitter_samples.strings('positive_tweets.json')
all_negative_tweets = twitter_samples.strings('negative_tweets.json')
```

```python
print('Number of postive tweets:', len(all_positive_tweets))
print('Number of negative tweets:', len(all_negative_tweets))

print(all_positive_tweets[0])
```

```python
# Declare a figure with a custom size
fig = plt.figure(figsize=(5, 5))

# labels for the two classes
labels = 'Positives', 'Negative'

# Sizes for each slide
sizes = [len(all_positive_tweets), len(all_negative_tweets)] 

# Declare pie chart, where the slices will be ordered and plotted counter-clockwise:
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

# Display the chart
plt.show()
```

```python
# preporcessing the tweet

tweet = all_positive_tweets[2277]
print(tweet)
```

```python
nltk.download('stopwords')
```

```python
import re 
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
```

```python
print(tweet)
# we have to remove the unwanted stuff like '#' , 'https'
```

```python
pre_tweet = re.sub(r'^RT[\s]+','',tweet)
```

```python
print(pre_tweet)
```

```python
pre_tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', pre_tweet)
```

```python
print(pre_tweet)
```

```python
pre_tweet = re.sub(r'#','',pre_tweet)
```

```python
print(pre_tweet)
```

```python
tokenizer = TweetTokenizer(preserve_case=False,strip_handles=True, reduce_len=True)
```

```python
tweet_tokens= tokenizer.tokenize(pre_tweet)
```

```python
print(tweet_tokens)
```

```python
stopwords_english = stopwords.words('english')
print(stopwords_english)
```

```python
print(string.punctuation)
```

```python
tweets_clean = []
for word in tweet_tokens:
    if(word not in stopwords_english and word not in string.punctuation):
        tweets_clean.append(word)
```

```python
print(tweets_clean)
```

```python
stemmer = PorterStemmer()
tweets_stem = []

for word in tweets_clean:
    stem_word = stemmer.stem(word)
    tweets_stem.append(stem_word)
```

```python
print(tweets_stem)
```
